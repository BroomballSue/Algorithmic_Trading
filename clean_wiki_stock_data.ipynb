```python
import pandas as pd
import h5py
```

```python
df=pd.read_csv("WIKI_PRICES.csv")

df.info(verbose=True)

df
```

```python
df["date"] = pd.to_datetime(df["date"])

df = df.set_index(['ticker', 'date']) # create a multiindex

df=df.drop(columns=['open', 'high'],axis=1)
df.drop(df.columns[[0, 1, 2, 3, 4]], axis = 1, inplace = True) # different ways of dropping columns

df=df[df.index.get_level_values(1)>'2000-01-01'] # because the date is part of a multiindex

df.columns = df.columns.str.replace('adj_', '') #renaming the remaining columns


df.info(verbose=True) # to check everything worked
df






```

```python
df.to_hdf('stock_data.h5', key='df', mode='w')  # writing the file to the disk in a hdf5 format
```
